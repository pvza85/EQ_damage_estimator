```python
import cv2
import data_preparation
import configparser
import pandas as pd
from data_preparation.converter import *
from data_preparation.building import *
import numpy as np

config = configparser.ConfigParser()
config.read('config.ini')
conf = config['DEFAULT']
convert = Converter(conf)
```

```python
data = pd.read_csv(conf['data_folder'] + conf['building_file'])
buildings = []
for index in data['ORIG_FID'].unique():
    Lng = data['X'][data['ORIG_FID'] == index]
    Lat = data['Y'][data['ORIG_FID'] == index]
    buildings.append(Building(Lng, Lat, index+1, convert, int(conf['window_size'])))
```

```python
height, width, c = convert.image.shape
window_size = int(conf['window_size'])
overlap = float(conf['overlap'])

df = pd.DataFrame(columns=('index', 'x0', 'y0', 'x1', 'y1', 'lng0', 'lat0', 'lng1', 'lat1', 'building_index', 'overlap'))
i = 0
for y in xrange(0, height-window_size, int(window_size * overlap)):
    for x in xrange(0, width-window_size, int(window_size * overlap)):
        
        for building in buildings:
            over = building.overlap(x, y, x + window_size, y + window_size)
            if over > float(conf['threshold']):
                break
        win_image = convert.image[y:y+window_size, x:x+window_size]
        means = np.average(win_image, axis=0)
        if means != (255, 255, 255):
            i += 1
            lng0, lat0 = convert.pix2coord(x, y)
            lng1, lat1 = convert.pix2coord(x+window_size, y+window_size)
            if over > float(conf['threshold']):
                bindex = building.index
            else:
                bindex = 0
            window = pd.DataFrame([{'index': int(i),  'x0':x, 'y0':y, 'x1':x + window_size, 'y1':y + window_size, \
                                    'lng0':lng0, 'lat0':lat0, 'lng1':lng1, 'lat1':lat1, 'building_index':bindex,\
                                    'overlap':over}])
            df = df.append(window)

            name = '{0}{1}/{2}.tif'.format(conf['data_folder'], conf['output_folder'], str(i).zfill(5))
            #print name
            cv2.imwrite(name,win_image)
            #break
    #break
df = df.set_index(['index'])
```
